```python
import pandas as pd
import numpy as np
```

```python
stock_name = 'RELIANCE.NS'
df = pd.read_csv('../data/sandbox/' + stock_name +'.csv')
print(df)
```

```python
df.dropna(inplace=True)
print(df)
```

```python
df['Signal'] = df['Adj Close'].rolling(20).mean() - df['Adj Close'].rolling(100).mean()
df['Position'] = (df['Signal'].apply(np.sign) + 1)/2
```
